```python
!pip install youtube-transcript-api

from urllib.parse import urlparse, parse_qs
from youtube_transcript_api import YouTubeTranscriptApi

def extract_video_id(url: str) -> str:
    """
    Extract the YouTube video ID from a URL.
    Raises ValueError if no 'v' parameter is found.
    """
    parsed = urlparse(url)
    qs = parse_qs(parsed.query)
    video_ids = qs.get('v')

    if not video_ids:
        raise ValueError(f"No video id found in URL: {url}")
    return video_ids[0]

if __name__ == "__main__":
    url = "https://www.youtube.com/watch?v=PZZz790YnzU"

    video_id = extract_video_id(url)

    api = YouTubeTranscriptApi()
    fetched = api.fetch(video_id, languages=['en']) # returns a FetchedTranscript

    text = "\n".join(snippet.text for snippet in fetched)

print(text)
```

```python
print(len(text))
```

```python
import math
def chunking(Stxt, st , end):
  return Stxt[st : end]

text1 = chunking(text,0,4000)
summerized = pipe(text1)
i=1

for i in range(int(math.ceil(len(text)/4000))):
  text1 = chunking(text,i*4000,(i+1)*4000)
  Nsummerized = pipe(text1)
  summerized[0]['summary_text'] += Nsummerized[0]['summary_text'] + " "

```
